```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_figure(data, title="Figure"):
    plt.figure(figsize=(6, 6))
    plt.plot(data[:, 0], data[:, 1], 'o-g', linewidth=2)
    plt.grid(True)
    plt.axhline(0, linewidth=1)
    plt.axvline(0, linewidth=1)
    plt.title(title)
    plt.xlim(-15, 15)
    plt.ylim(-15, 15)
    plt.show()
```

```python
def plot_two_figures(data1, data2, titles=("First", "Second")):
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    for ax, data, title in zip(axes, [data1, data2], titles):
        ax.plot(data[:, 0], data[:, 1], 'o-g', linewidth=2)
        ax.set(title=title, xlim=(-15, 15), ylim=(-15, 15))
        ax.grid(True)
        ax.axhline(0, color='black', linewidth=1)
        ax.axvline(0, color='black', linewidth=1)

    plt.tight_layout()
    plt.show()
```

```python
def plot_images_in_row(number_images, images, titles=None, figsize=(15, 5)):
    fig, axes = plt.subplots(1, number_images, figsize=figsize)

    if number_images == 1:
        axes = [axes]

    for i in range(number_images):
        axes[i].imshow(images[i])

        if titles is not None and i < len(titles):
            axes[i].set_title(titles[i])

        axes[i].axis('off')

    plt.tight_layout()
    plt.show()
```

## Операції над фігурами

```python
points = np.array([
    [2, 2], [3, 6], [7, 2], [7, -2], [4, -2], [4, -4], [-2, -3], [-3, -5], [-5, -4], [-3, 4], [-1, 2], [2, 2]
])

plot_figure(points, "Початкова фігура")
```

```python
# Підготовка даних: транспонуємо точки (2, N)
points_T = points.T

# Додаємо рядок одиниць для створення однорідних координат (3, N)
points_h = np.vstack([points_T, np.ones(points_T.shape[1])])

# points, points_h
```

Зсув (Translation)
$$\begin{bmatrix} 1 & 0 & Δx \\ 0 & 1 & Δy \\ 0 & 0 & 1 \end{bmatrix}$$

Масштабування (Scaling)
$$\begin{bmatrix} k_x & 0 & 0 \\ 0 & k_y & 0 \\ 0 & 0 & 1 \end{bmatrix}$$

Поворот (Rotation)
$$\begin{bmatrix} \cos\theta & -\sin\theta & 0 \\ \sin\theta & \cos\theta & 0 \\ 0 & 0 & 1 \end{bmatrix}$$

Дзеркалення (Reflection)
$$\begin{bmatrix} ±1 & 0 & 0 \\ 0 & ±1 & 0 \\ 0 & 0 & 1 \end{bmatrix}$$

Відображення зсуву (Shearing / Skewing)
$$\begin{bmatrix} 1 & c_x & 0 \\ c_y & 1 & 0 \\ 0 & 0 & 1 \end{bmatrix}$$

```python
# Зсув (translation) на dx=2, dy=3
M_trans = np.array([
    [1, 0, 2],
    [0, 1, 3],
    [0, 0, 1]
])

# Масштабування (scaling) у 1.5 рази
M_scale = np.array([
    [1.5, 0, 0],
    [0, 1.5, 0],
    [0, 0, 1]
])

# Поворот (rotation) на 90 градусів
theta = np.radians(90)
M_rot = np.array([
    [np.cos(theta), -np.sin(theta), 0],
    [np.sin(theta),  np.cos(theta), 0],
    [0, 0, 1]
])

# Дзеркалення (reflection) відносно осі Y
M_reflect = np.array([
    [-1, 0, 0],
    [ 0, 1, 0],
    [ 0, 0, 1]
])

# Відображення зсуву (shear) - нахил
M_shear = np.array([
    [1, 1.5, 0],
    [0, 1, 0],
    [0, 0, 1]
])
```

```python
result = M_trans @ points_h
# result = M_scale @ points_h
# result = M_rot @ points_h
# result = M_shear @ points_h

transformed_points = result.T[:, :2]

plot_two_figures(points, transformed_points, ("Початкова фігура", "Трансформована фігура"))
```

## Перспективне перетворення

## Зміна яскравості та контрасту зображення

```python
img = cv2.imread("input/photo.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

Зміна яскравості $$I^*(x, y) = I(x, y) + \beta$$

```python
beta = 100

beta_matrix = np.full(img.shape, beta, dtype=np.uint8)

img_bright = cv2.add(img, beta_matrix)
img_dark = cv2.subtract(img, beta_matrix)


titles = ["Оригінал", f"Яскравіше (+{beta})", f"Темніше (-{beta})"]
images = [img, img_bright, img_dark]

plot_images_in_row(3, images, titles, figsize=(15, 5))
```

Зміна контрасту $$I^*(x, y) = I(x, y) \cdot \alpha$$

```python
# Конвертуємо в float32, щоб при множенні не обрізало значення
img_float = img.astype(np.float32)

alpha_high = 2
alpha_low = 0.5


img_high = np.clip(img_float * alpha_high, 0, 255).astype(np.uint8)
img_low = np.clip(img_float * alpha_low, 0, 255).astype(np.uint8)


titles = ["Оригінал", f"Контрастність (alpha={alpha_high})", f"Контрастність (alpha={alpha_low})"]
images = [img, img_high, img_low]

plot_images_in_row(3, images, titles, figsize=(15, 5))
```

$$I^*(x, y) = I(x, y) \cdot \alpha + \beta$$

```python
alpha = 1.8
beta = 50

result = np.clip(alpha * img.astype(np.float32) + beta, 0, 255).astype(np.uint8)


titles = ["Оригінал", f"Результат (alpha={alpha}, beta={beta})"]
images = [img, result]

plot_images_in_row(2, images, titles, figsize=(10, 5))
```

## Морфологічні операції

Erosion, Dilation, Opening, Closing

```python
j_img = cv2.imread("input/j.webp")

kernel = np.ones((5,5),np.uint8)

erosion = cv2.erode(j_img,kernel,iterations = 1)
dilation = cv2.dilate(j_img,kernel,iterations = 1)

# plt.imshow(j_img)
plot_images_in_row(3, [j_img, erosion, dilation], ["Original", "Erosion", "Dilation"], figsize=(15, 5))
```

```python
j_open_img = cv2.imread("input/j_for_opening.png")

kernel = np.ones((7,7),np.uint8)

opening = cv2.morphologyEx(j_open_img, cv2.MORPH_OPEN, kernel)

plot_images_in_row(2, [j_open_img, opening], ["Original", "Opening"], figsize=(10, 5))
```

```python
j_close_img = cv2.imread("input/j_for_closing.png")

kernel = np.ones((7,7),np.uint8)

closing = cv2.morphologyEx(j_close_img, cv2.MORPH_CLOSE, kernel)

plot_images_in_row(2, [j_close_img, closing], ["Original", "Closing"], figsize=(10, 5))
```

## Приклад вирівнювання тексту

```python
book_img = cv2.imread("input/rotated_page.png")
book_img = cv2.cvtColor(book_img, cv2.COLOR_BGR2RGB)

plt.imshow(book_img)
```

```python
# 1. Завантаження та перетворення у ч/б
gray = cv2.cvtColor(book_img, cv2.COLOR_BGR2GRAY)
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV) # інвертуємо, щоб текст був білим

# 2. Підрахунок білих пікселів у кожному рядку (по ширині)
# binary == 255 дає масив True/False, .sum(axis=1) рахує True в кожному рядку
hist = np.sum(binary == 255, axis=1)


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 12), gridspec_kw={'width_ratios': [3, 1]})

ax1.imshow(binary, cmap='gray')
ax1.set_title("Бінарне зображення")
ax1.axis('off')

y_pos = np.arange(len(hist))
ax2.fill_betweenx(y_pos, hist, color='blue', alpha=0.8, edgecolor='blue')
ax2.plot(hist, y_pos, color='blue', linewidth=1)

ax2.set_title("Кількість пікселів (профіль)")
ax2.set_ylim(len(hist), 0)
ax2.set_xlabel("Кількість білих пікселів")
ax2.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

```python
def get_score(image, angle):
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, angle, 1.0)
    rotated = cv2.warpAffine(image, M, (w, h), flags=cv2.INTER_CUBIC)

    # Обчислення гістограми (як у вашому коді)
    hist = np.sum(rotated == 255, axis=1)

    # Повертаємо стандартне відхилення (або дисперсію)
    return np.std(hist)
```

```python
best_angle = 0
max_std = 0

for angle in np.arange(-15, 0, 0.1):
    current_std = get_score(binary, angle)
    if current_std > max_std:
        max_std = current_std
        best_angle = angle

print(f"Оптимальний кут повороту: {best_angle}")
```

```python
(h, w) = binary.shape[:2]
M = cv2.getRotationMatrix2D((w // 2, h // 2), best_angle, 1.0)
final_corrected = cv2.warpAffine(binary, M, (w, h), flags=cv2.INTER_CUBIC)

# plt.imshow(final_corrected)
```

```python
hist = np.sum(final_corrected == 255, axis=1)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 12), gridspec_kw={'width_ratios': [3, 1]})

ax1.imshow(final_corrected, cmap='gray')
ax1.set_title("Бінарне зображення")
ax1.axis('off')

y_pos = np.arange(len(hist))
ax2.fill_betweenx(y_pos, hist, color='blue', alpha=0.8, edgecolor='blue')
ax2.plot(hist, y_pos, color='blue', linewidth=1)

ax2.set_title("Кількість пікселів (профіль)")
ax2.set_ylim(len(hist), 0)
ax2.set_xlabel("Кількість білих пікселів")
ax2.grid(True, linestyle='--', alpha=0.5)

plt.show()
```
